```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.random.normal(10, 5, 100)
x = x.astype(int)
print(x)
```

```python
sns.distplot(x, norm_hist=True, hist=True)
plt.title("PDF")
plt.show()
```

```python
mean = x.mean()
sd = x.std()

print("mean is :", mean, " sd is : ", sd)
y = [((v-mean)/sd) for v in x]
sns.distplot(y)
plt.title("Z Score")
plt.show()
```

```python
sns.distplot(x, hist=False, kde_kws={"cumulative":True})
plt.title("CDF")
plt.show()
```

# What is Central Limit Theorem
    1. Population mean = Sample Distribution mean
    2. error = satndard deviation of Population / sqrt (nop of samples)
    

```python
x = np.random.randint(0, 100, 10000, int)
print("Mean of x is :", x.mean())
sns.distplot(x)
```

```python
sample = [np.random.choice(x) for v in range(100)]
print("Sample Mean is : ", np.mean(sample))
sns.distplot(sample)
```

```python
# Sample Distributions
samples = []

for i in range(50):
    s = [np.random.choice(x) for v in range(100)]
    samples.append(np.mean(s))

print("Samples Means is : ", np.mean(samples))
sns.distplot(samples)
```
